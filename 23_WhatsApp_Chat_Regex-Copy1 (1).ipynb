```python
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
f=open("WhatsApp Chat with Mysql kumood wala 1pm.txt",encoding="utf-8") ##f - file pointer  ## it opens the file in read mode
```

```python
data=f.read()
data
```

  #### Seggregate data and messages into columns where columns contains user name as well .m

```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages=re.split(pattern,data)
messages
```

```python
messages=messages[1:]
messages
```

```python
dates=re.findall(pattern,data)
dates
```

```python
df=pd.DataFrame({"user_message":messages,"message_date":dates})
df
```

```python
df.info()
```

#### Since date is object so convert it into datetime.

```python
df["message_date"]=pd.to_datetime(df["message_date"],format='%d/%m/%Y, %H:%M - ')
df["message_date"]
```

```python
df["message_date"].dtypes
```

```python
df["year"]=df["message_date"].dt.year
df["month"]=df["message_date"].dt.month_name()
df["day"]=df["message_date"].dt.day
df["hour"]=df["message_date"].dt.hour
df["hour"]=df["message_date"].dt.month
df["day_name"]=df["message_date"].dt.day_name()
df["Date"]=df["message_date"].dt.date
df.head()
```

```python
df.drop(columns="message_date",inplace=True)
df.head()
```

```python
users=[]
messages=[]

for i in df["user_message"]:
    print(i)
    if "changed the subject" not in i:
        li=i.split(":",maxsplit=1)
    else:
        li=[]
        li.append(i)
    print(li)
    if len(li)>=2:
        users.append(li[0])
        messages.append(li[1])
    else:
        users.append("group notification")
        messages.append(li[0])
df["users"]=users
df["messages"]=messages
```
